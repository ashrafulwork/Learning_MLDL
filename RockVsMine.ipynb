```python
X = sonar_data.drop(columns = 'R', axis = 1 )
Y = sonar_data['R']
```

```python
X.head()
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.3, stratify = Y, random_state=50)
```

```python
print(X_train)
```

```python
print(Y_train)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
 X_train_prediction = model.predict(X_train)
 training_data_accuracy = accuracy_score(X_train_prediction, Y_train)

 print("Accuracy of our training data : ", training_data_accuracy)
```

```python
 X_test_prediction = model.predict(X_test)
 test_data_accuracy = accuracy_score(X_test_prediction, Y_test)

 print("Accuracy of our testing data : ", test_data_accuracy)
```
