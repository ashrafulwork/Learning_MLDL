Creating a pandas dataframe

```python
# Manually load the Boston Housing dataset
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)

# Create the feature matrix
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])

# Extract the target variable (median home prices)
target = raw_df.values[1::2, 2]

# Mimicking the load_boston() output structure
boston_dataset = {
    'data': data,
    'target': target,
    'feature_names': np.array([
        'CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
        'PTRATIO', 'B', 'LSTAT'
    ]),
    'DESCR': 'Boston Housing Dataset\nThis dataset contains 13 features related to housing prices in Boston. The target variable is the median value of owner-occupied homes in $1000s.'
}
```

```python
print(boston_dataset)
```

```python
# pandas DataFrame

# boston_df = pd.DataFrame(boston_dataset.data, columns = boston_dataset.feature_names)
boston_df = pd.DataFrame(boston_dataset['data'], columns=boston_dataset['feature_names'])
```
